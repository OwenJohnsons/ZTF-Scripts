```python
# get rid of pandas indexing and sort into lists
os.chdir(r'C:\Users\laura\OneDrive\Documents\PhD\y1\search_dat\ztf_100_transients\moniter')
#sort out dates
d = datetime.today()
date = datetime.today().strftime('%d_%m_%Y')
y = (d - timedelta(days=1))
yday = y.strftime('%d_%m_%Y')

t = pd.read_csv('{}.csv'.format(yday))

n = t['Name']
names = []
probs = []
cla = []
for i in range(0,len(n)):
    names.append(n[i])
 
for i in range(0,len(n)):

    xxx = alerce.query_objects(oid = names[i],classifier="lc_classifier_transient",format="pandas")

    probs.append((xxx['probability'][0]))
    cla.append((xxx['class'])[0])



# for j in range(0,len(n)):
#     detections = alerce.query_detections(n[j],format="json")
#     print(len((detections[j]).get('magpsf')))
```

```python
#initialise all parameters to be initialised
n_g = []
n_r = []
for i in names:
    count_g = 0
    count_r = 0
    detections = alerce.query_detections(i,format="json")
    for j in range(0,len(detections)):
        fil_n = (detections[j]).get('fid')
        if fil_n == 1:
            count_g += 1
       
        if fil_n == 2:
            count_r +=1
    n_g.append(count_g)
    n_r.append(count_r)

print(len(n_g))
```
